### Bir Metin Dosyası Okuyun 

```python
import collections
hamlet=open("hamlet_TXT_FolgerShakespeare.txt").read()
```

```python
hamlet
```

```python
kelime_listesi=hamlet.split() # kelime listesi oluşturduk
```

```python
word_freq=collections.Counter(kelime_listesi)  # kelimeleri tek tek saydık
```

```python
word_freq.most_common(10) # en fazla kullanılan 10 kelimeyi sıraladık
```
